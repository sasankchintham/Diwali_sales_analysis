```python
plt.bar('Male', mc, color='blue', width=0.4, label='Male')
plt.bar('Female', fc, color='pink', width=0.4, label='Female')

plt.title('Count of Males and Females in Gender Column')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend()
plt.text('Male', mc ,str(mc), fontsize=10, color='black')
plt.text('Female', fc , str(fc),fontsize=10, color='black')
plt.show()
```

### How is the count of each category in the 'Gender' column visualized?

### male-3409, female-7842

### pie chart

```python
labels = ['Male', 'Female']
sizes = [mc, fc]
colors = ['blue', 'pink']
plt.pie(sizes,labels=labels,colors=colors)
plt.title('Gender count')
plt.show()
```

#### The above is the pie chart representing the genders(blue-male,pink-female)
